```python
processes_2020 = ['Airing_home','Airing_home_end','cook_and_eat','do_the_dishes','do_walk','drink','eat_cold','eat_warm','exercise','go_work','prepared_sleep','rest','shop','someone_at_entrace','use_computer','watch_tv','wc']
```

```python
def delete_noises(df):
    events_to_drop = []
    
    for i in df.index:
        concept_name = df.at[i, 'concept:name']
        if concept_name.startswith("noise"):
            events_to_drop.append(i)
            print(f"Deleting event noise {concept_name}")

    df = df.drop(df.index[events_to_drop])
    
    return df
```

```python
event_log_df = delete_noises(event_log_df)
```

```python
event_log_df.reset_index(inplace=True)
```

```python
def split_by_process(df, process):
    events_to_drop = []
    
    for i in df.index:
        case_concept_name = df.at[i, 'reference:case:concept:name']
        if case_concept_name != process:
            events_to_drop.append(i)
            print(f"Deleting process {case_concept_name} for event {df.at[i, 'concept:name']}")

    df = df.drop(df.index[events_to_drop])
    
    return df

    
```

```python
from conversion import write_xes
```

```python
id = 0
for process in processes_2020:
    id = id + 1
    print(f"{id} - Going to split at 3 for process {process}")
    process_df = split_by_process(event_log_df, process)
    output_file = f'abstraction/dataset-3am/processes/split-3am-splitted_at_3-{id}-{process}.xes'
    write_xes(process_df, output_file)
```
